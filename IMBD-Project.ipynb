```python
#20. Classify Movies Based on Ratings [Excellent, Good, and Average]
data.columns
```

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
    
```

```python
#22. Find Unique Values From Genre
data.columns
```

```python
data['Genre']
```

```python
list1=[]
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```
